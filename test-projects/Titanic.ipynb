```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16 :
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels= []

for level in deck:
    levels.append(level[0])
```

```python
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.catplot('Cabin',kind='count',data=cabin_df,palette='winter_d')
```

```python
sns.catplot('Cabin',kind='count',data=cabin_df,palette='winter_d',order = 'ABCDEFGT')
```

```python
cabin_df = cabin_df[cabin_df.Cabin !='T']
sns.catplot('Cabin',kind='count',data=cabin_df,palette='summer')
```

```python
sns.catplot('Embarked',kind='count',data=titanic_df,hue='Pclass',order='CQS')
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] >0] ='With Family'

titanic_df['Alone'].loc[titanic_df['Alone'] ==0] ='Alone'
```

```python
titanic_df.head()
```

```python
sns.catplot('Alone',kind='count',data=titanic_df,palette='Blues')
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})

sns.catplot('Survivor',kind='count',data=titanic_df,palette='Set1')
```

```python
sns.catplot('Pclass','Survived',hue='person',kind='point',data=titanic_df)
```
